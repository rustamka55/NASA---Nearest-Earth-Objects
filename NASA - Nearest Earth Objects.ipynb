```python
X = df[["relative_velocity","est_diameter_max","miss_distance","absolute_magnitude"]]
y = df["hazardous"].astype(int)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1,random_state=42,test_size=0.2)
```

```python
for train_index, test_index in split.split(X,y):
    X_train, X_test = X.loc[train_index],X.loc[test_index]
    y_train, y_test = y.loc[train_index],y.loc[test_index]
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint
from scipy.stats import uniform as sp_randfloat
```

```python
tree_clf = Pipeline([
    ('standard_scaler', StandardScaler()),
    ('clf',DecisionTreeClassifier(random_state=42))
])
```

```python
params = {
    "clf__max_depth":sp_randint(4, 50)
}
```

```python
grid_pipeline = RandomizedSearchCV(tree_clf,params,scoring='recall',random_state=42)
```

```python
grid_pipeline.fit(X_train,y_train)
```

```python
grid_pipeline.best_estimator_
```

```python
pred = grid_pipeline.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
```

```python
forest_clf = Pipeline([
    ('standard_scaler', StandardScaler()),
    ('clf',RandomForestClassifier(random_state=42))
])
```

```python
params = {
    "clf__n_estimators":sp_randint(50, 500),
    "clf__max_depth":sp_randint(4, 50)
}
```

```python
grid_pipeline = RandomizedSearchCV(forest_clf,params,scoring='f1',random_state=42)
```

```python
grid_pipeline.fit(X_train,y_train)
```

```python
grid_pipeline.best_estimator_
```

```python
pred = grid_pipeline.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```
